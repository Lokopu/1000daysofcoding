```python
import pandas as pd


df = pd.read_json("flux-velos-et-trottinettes-en-temps-reel-capteurs-geres-par-bruxelles-mobilite.json")
df.head(3)
```

```python
import altair as alt

# Basic bar chart
alt.Chart(df).mark_bar().encode(
    alt.X("commune").title("Commune"),
    alt.Y("data_year_cnt").title("Count of cyclists (year)")
)
```

```python
# Bar chart with legend and title
alt.Chart(df, title="Count of yearly cyclists by adresse and commune").mark_bar().encode(
    alt.X("adresse_nl").title("Address (NL)"),
    alt.Y("data_year_cnt").title("Count of cyclists (year)"),
    color=alt.Color("commune").title("Commune")
)
```

```python
input_dropdown = alt.binding_select(options=df["commune"].unique(), name="Commune ")

selection = alt.selection_point(fields=['commune'], bind=input_dropdown)

color = (
    alt.when(selection)
    .then(alt.Color("commune:N")
    .title("Commune")
    .legend(None))
    .otherwise(alt.value("lightgray"))
)

alt.Chart(df).mark_bar().encode(
    x='adresse_nl:N',
    y="data_year_cnt:Q",
    color=color,
).add_params(
    selection
)
```
